```python
import pandas as pd

# ---------------------------------------------
#                Load dataset
# ---------------------------------------------

path = "golf_dataset/golf_dataset_wide_format.csv"
all_data = pd.read_csv(path)
```

```python
from sklearn.model_selection import train_test_split

# split data into training (90%) and test (10%)
training_data, test_data = train_test_split(
    all_data,
    test_size=0.1,
    random_state=42,
    shuffle=True
)

print(f"Length of all data: {len(all_data)}")
print(f"Length of training data: {len(training_data)}")
print(f"Length of testing data: {len(test_data)}")
```

```python
# save to csv's
all_data.to_csv("golf_dataset/split_sets/all_data.csv", index=False)
training_data.to_csv("golf_dataset/split_sets/training_data.csv", index=False)
test_data.to_csv("golf_dataset/split_sets/test_data.csv", index=False)
```
